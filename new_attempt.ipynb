```python
import numpy as np
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('input_game.csv')

df['p1_action'] = df['p1_action'].replace({'TRUST': 0, 'CHEAT': 1})
df['p2_action'] = df['p2_action'].replace({'TRUST': 0, 'CHEAT': 1})

df
```

```python
def longest_common_prefix(*series):
    if not series:
        return ""

    prefix = series[0]
    for s in series[1:]:
        i = 0
        while i < len(prefix) and i < len(s) and prefix.iloc[i] == s.iloc[i]:
            i += 1
        prefix = prefix.iloc[:i]

    return prefix

player_prefix_dict = {}

for player_id in range(201):
    
    lcp = pd.Series()
    temp = df[df['p1_id'] == player_id]

    for game_id in range(3014):
        game_data1 = temp[temp['game_id'] == game_id]

        if(game_data1.empty):
            continue

        if(lcp.empty):
            lcp = game_data1['p1_action']
        else:
            lcp = longest_common_prefix(lcp, game_data1['p1_action'])

    temp = df[df['p2_id'] == player_id]
    for game_id in range(3014):
        game_data2 = temp[temp['game_id'] == game_id]

        if(game_data1.empty):
            continue

        if(lcp.empty):
            lcp = game_data1['p2_action']
        else:
            lcp = longest_common_prefix(lcp, game_data1['p2_action'])


    player_prefix_dict[player_id] = lcp
   

print(player_prefix_dict)   
```
